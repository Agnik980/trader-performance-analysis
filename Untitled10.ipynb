# **PART C**

# **Strategy 1: Risk Control in Fear Markets**

During Fear periods, traders tend to have lower PnL and higher losses.
It is recommended to reduce leverage and trade with smaller position sizes to minimize risk.
# **Strategy 2: Avoid Overtrading in Greed Markets**

In Greed conditions, traders increase trade frequency and leverage.
Traders should avoid overtrading and consider booking profits early to protect gains.
## **Strategy 3: Segment-Based Strategy **

High leverage traders show higher volatility in returns.
It is advisable for high leverage traders to trade selectively during stable (Greed) periods and avoid Fear phases.
